```python
import pandas as pd
import plotly.express as px
from sklearn.preprocessing import MinMaxScaler

df = pd.read_csv('../src/data/processed/daily_metrics.csv', parse_dates=['Unnamed: 0'])
df.rename(columns={'Unnamed: 0': 'date'}, inplace=True)

scaler = MinMaxScaler()
normalized_values = scaler.fit_transform(df.iloc[:, 1:])  

normalized_df = pd.DataFrame(normalized_values, columns=df.columns[1:])
normalized_df['date'] = df['date']

normalized_melted = normalized_df.melt(id_vars=['date'], var_name='Metric', value_name='Normalized Value')

fig = px.line(normalized_melted, x='date', y='Normalized Value', color='Metric',
              title='Normalized Metrics Over Time',
              labels={'date': 'Date', 'Normalized Value': 'Normalized Value'})
fig.update_layout(width=1400, height=600, xaxis_title="Date", yaxis_title="Normalized Value")
fig.update_xaxes(tickangle=45) 
fig.show()
```

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv('../src/data/processed/daily_metrics.csv', parse_dates=['Unnamed: 0'])
df.rename(columns={'Unnamed: 0': 'date'}, inplace=True)

demand_metrics = [col for col in df.columns if 'demand' in col]
price_metrics = [col for col in df.columns if 'price' in col]
wind_metrics = [col for col in df.columns if 'wind' in col]
solar_metrics = [col for col in df.columns if 'solar' in col]

groups = {
    'Demand': demand_metrics,
    'Price': price_metrics,
    'Wind': wind_metrics,
    'Solar': solar_metrics
}

for group_name, metrics in groups.items():
    if not metrics:  # Skip empty groups
        continue
    
    df_melted = df.melt(id_vars=['date'], value_vars=metrics, var_name='Metric', value_name='Value')

    fig = px.line(df_melted, x='date', y='Value', color='Metric',
                  title=f'{group_name} Metrics Over Time',
                  labels={'date': 'Date', 'Value': 'Value'},
                  width=1400, height=600)

    fig.update_xaxes(tickangle=45)

    fig.show()
```
